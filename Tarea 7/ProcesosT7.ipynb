```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

![Alt text](image.png)

```python
def grafo_anillo(N):
    # Crear un grafo vac√≠o
    G = nx.Graph()
    Nodos = [i for i in range(1,N)]
    G.add_nodes_from(Nodos)
    pesos = []
    for i in range(1,N):
        pesos.append((i,i+1,1/2))
    pesos.append((N,1,1/2))
    
    G.add_weighted_edges_from(pesos)

    return G

def caminata_al(N,pos,n,p):
    G = grafo_anillo(N)
    d = dict(zip(list(G.nodes()),[0 for i in range(N)]))
    
    for i in range(n):
        if random.random() < p:
            d[list(nx.neighbors(G,pos))[1]] += 1
            pos = list(nx.neighbors(G,pos))[1]
        else:
            d[list(nx.neighbors(G,pos))[0]] += 1        
            pos = list(nx.neighbors(G,pos))[0]
        
    return d
    
        
```

```python
d = caminata_al(10,1,1000000,0.5)
nodos = list(d.keys())
visitas = [d[i] for i in range(1,len(nodos)+1)]

plt.bar(nodos,visitas)
plt.show()
d
```
